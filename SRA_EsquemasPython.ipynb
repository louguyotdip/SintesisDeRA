```python
with schemdraw.Drawing() as d:
    d += (op := elm.Opamp(leads=True).label('LM324', loc='bot'))
    d += elm.Line().down(0.5).at(op.in2) # Ground connection
    d += elm.Ground(lead=False)

    # Input Resistors - defining points using coordinates relative to op.in1
    d += (Rin1 := elm.Resistor().at(op.in1).left().idot().label('$1k\Omega$').label(r'$V_{2}$', loc='left')) # Used raw string
    d += (lin := elm.Line().up(1).at(op.in1))

    # Feedback Resistor
    d += (Rf := elm.Resistor().tox(op.out).label('$30k\Omega$'))
    d += elm.Line().toy(op.out).dot() # Dot at the output/feedback junction

    # Adjusting placement for Rin2
    d += (Rin2 := elm.Resistor().at(Rf.start).left().label('$1k\Omega$').label(r'$V_{1}$', loc='left')) # Used raw string
    d += (lin := elm.Line().up(1).at(op.in1))


    # Output
    d += elm.Line().right(0.5).at(op.out).label(r'$V_{O}$', loc='right') # Used raw string
```
