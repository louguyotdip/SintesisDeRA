## Topologías bicuadráticas

**Topología de realimentación negativa**

Función de transferencia feedforward
: $$ T_{FF}=\left.\frac{V^-}{V_{in}}\right|_{V_{o}=0} $$

Función de transferencia feedback: $$ T_{FB}=\left.\frac{V^-}{V_o}\right|_{V_{in}=0} $$



Analizando el circuito, la tensión de salida está dada por $$V_o =(V^+ -V^-)A $$


por supersposición $$ V^- = T_{FF}V_{in} + T_{FB}V_o $$
Notar que $$V^+ =0 $$
tenemos $$ V_o = -V^-A_d = -(T_{FF}V_{in} + T_{FB}V_o)A_d $$
entonces
 $$ A_v = \frac{V_o}{V_{in}} = -\frac{T_{FF}}{T_{FB}+1/A_d} $$



 luego $$ \frac{V_o}{V_{in}}= -\frac{T_{FF}}{T_{FB}} $$
 $$ A_v= -\frac{N_{FF
 }}{N_{FB}} $$

**Topología de realimentación positiva**

Función de transferencia feedforward
: $$ T_{FF}=\left.\frac{V^+}{V_{in}}\right|_{V_{o}=0} $$

Función de transferencia feedback: $$ T_{FB}=\left.\frac{V^+}{V_o}\right|_{V_{in}=0} $$



Analizando el circuito, la tensión de salida está dada por $$V_o =(V^+ -V^-)A $$


donde $$ V^- = \frac{V_or_1}{r_1+(k-1)r_1}= V_o/k $$
y $$V^+ = V_{in}T_{FF} +V_oT_{FB} $$
entonces $$ V_o = (V_{in}T_{FF} +V_oT_{FB}-\frac{V_o}{k})A_d $$
entonces
 $$ A_v = \frac{V_o}{V_{in}} = \frac{kT_{FF}}{1-kT_{FB}+k/A_d} $$



suponiendo que el op amp es ideal, $$A_d = \infty $$
$$A_v=\frac{kT_{FF}}{1-kT_{FB}} $$

Como $$T_{FF}$$ y $$T_{FB}$$ tienen los mismos polos, la función de transferencia resulta
$$A_v=\frac{kN_{FF}}{D-kN_{FB}}$$
